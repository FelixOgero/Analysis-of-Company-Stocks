```python
JPM_Stock["Adj Close"].plot(figsize=(10,7))
plt.title("Adjusted Closing Prices")
plt.xlabel("Date")
plt.ylabel("Adj Close")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Linear Regression to Predict JP Morgan Stock Prices

# Selecting features (Open, High, Low) and target variable (Close)
features = JPM_Stock[['Open', 'High', 'Low']]
target = JPM_Stock['Close']
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Creating a linear regression model
model = LinearRegression()

# Training the model
model.fit(X_train, y_train)
```

```python
# Making predictions on the test set
y_pred = model.predict(X_test)
print(y_pred)
print(y_test)
```

```python
# Plotting actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test.index, y_test, color='blue', label='Actual Values')
plt.scatter(y_test.index, y_pred, color='red', label='Predicted Values')
plt.title('Linear Regression: Actual vs Predicted Values')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
# Calculate the accuracy of the model
from sklearn import metrics
accuracy = metrics.r2_score(y_test, y_pred)
print(f'Accuracy of the model: {accuracy * 100:.2f}%')
```

```python
accuracy = model.score(X_test, y_test)
print(f'Accuracy of the model: {accuracy * 100:.2f}%')
```
